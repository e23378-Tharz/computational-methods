```python
import numpy as np
import os

data = np.array([
    [7, 20.0, 45.0],
    [8, 20.5, 45.4],
    [9, 21.0, 45.8],
    [10, 21.4, 46.2],
    [11, 21.8, 46.6],
    [12, 22.1, 47.0],
    [13, 22.3, 47.3],
    [14, 22.5, 47.6],
])

np.save("plane.npy", data)
print("Saved to:", os.path.abspath("plane.npy"))
```

```python
import numpy as np

data = np.load("plane.npy")

print(data)
```

```python
t = data[:,0]
alpha_deg = data[:,1]
beta_deg = data[:,2]

print("Time =", t)
print("Alpha =", alpha_deg)
print("Beta =", beta_deg)
```

```python
alpha = np.deg2rad(alpha_deg)
beta = np.deg2rad(beta_deg)

print(alpha)
print(beta)
```

```python
a = 500

x = a * np.tan(beta) / (np.tan(beta) - np.tan(alpha))

y = a * np.tan(beta) * np.tan(alpha) / (np.tan(beta) - np.tan(alpha))

print("x =", x)
print("y =", y)
```

```python
h = 1

x_prime = np.zeros(len(x))
y_prime = np.zeros(len(y))

# Forward difference
x_prime[0] = (x[1] - x[0]) / h
y_prime[0] = (y[1] - y[0]) / h

# Centered difference
for i in range(1, len(x)-1):
    x_prime[i] = (x[i+1] - x[i-1]) / (2*h)
    y_prime[i] = (y[i+1] - y[i-1]) / (2*h)

# Backward difference
x_prime[-1] = (x[-1] - x[-2]) / h
y_prime[-1] = (y[-1] - y[-2]) / h

print("x' =", x_prime)
print("y' =", y_prime)
```

```python
v = np.sqrt(x_prime**2 + y_prime**2)

print("Speed =", v)
```

```python
speed_change = np.diff(v)

largest_drop_index = np.argmin(speed_change)

print("Largest decrease occurs between:")
print(t[largest_drop_index], "and", t[largest_drop_index+1])
```
