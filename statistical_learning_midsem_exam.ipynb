## Q A1 — Probability Space `[4 marks]`

Define a probability space $(\Omega, \mathfrak{U}, P)$. State the three axioms (Kolmogorov) that $P$ must satisfy.

### ✅ Model Answer

A probability space is a triple $(\Omega, \mathfrak{U}, P)$ where:
- $\Omega$ = **sample space** (set of all possible outcomes)
- $\mathfrak{U}$ = **$\sigma$-algebra** on $\Omega$ (collection of measurable events)
- $P: \mathfrak{U} \to [0,1]$ = **probability measure**

**The three Kolmogorov axioms:**

1. **Non-negativity:** $P(A) \geq 0$ for all $A \in \mathfrak{U}$
2. **Normalisation:** $P(\Omega) = 1$
3. **Countable additivity:** For pairwise disjoint $A_1, A_2, \dots \in \mathfrak{U}$:
$$P\left(\bigcup_{i=1}^{\infty} A_i\right) = \sum_{i=1}^{\infty} P(A_i)$$

> **Marking:** 1 mark: correct triple definition. 1 mark each for the three axioms.

## Q A2 — Random Variable & Density `[4 marks]`

Let $X: \Omega \to \mathbb{R}$ be a random variable with density $f_X$. State the Radon-Nikodym condition under which $f_X$ exists, and write the formula connecting $f_X$ to $P_X$ (the law of $X$).

### ✅ Model Answer

**Condition:** The law $P_X$ must be **absolutely continuous** with respect to Lebesgue measure $\lambda$, written $P_X \ll \lambda$. This means: if $\lambda(B) = 0$ then $P_X(B) = 0$.

By the **Radon-Nikodym theorem**, there exists a unique (a.e.) non-negative measurable function $f_X$ (the density) such that:

$$f_X = \frac{dP_X}{d\lambda} \qquad \text{(Radon-Nikodym derivative)}$$

$$P(X \in B) = P_X(B) = \int_B f_X(x)\, dx \qquad \forall \text{ Borel } B$$

$$\int_{-\infty}^{\infty} f_X(x)\, dx = 1$$

> **Marking:** 1 mark: absolute continuity condition. 1 mark: Radon-Nikodym statement. 1 mark: integral formula. 1 mark: normalisation.

## Q A3 — Conditional Expectation `[4 marks]`

State the two defining properties of the conditional expectation $\mathbb{E}[X \mid Y]$. Why does the Doob-Dynkin lemma imply that $\mathbb{E}[X \mid Y]$ must be a function of $Y$?

### ✅ Model Answer

$\mathbb{E}[X \mid Y]$ is the unique (a.s.) random variable $Z$ satisfying:

1. **Measurability:** $Z$ is $\sigma(Y)$-measurable
2. **Partial averaging:** For every $G \in \sigma(Y)$:
$$\int_G Z\, dP = \int_G X\, dP$$

**Doob-Dynkin connection:** The Doob-Dynkin lemma states that a random variable $Z$ is $\sigma(Y)$-measurable **if and only if** $Z = g(Y)$ for some measurable function $g$.

Since property (i) forces $Z$ to be $\sigma(Y)$-measurable, Doob-Dynkin immediately gives $Z = g(Y)$ — the conditional expectation is entirely determined by the value of $Y$.

> **Marking:** 1 mark each for the two properties. 1 mark: Doob-Dynkin statement. 1 mark: connecting it to the conclusion $Z = g(Y)$.

## Q A4 — MLE & Bessel's Correction `[4 marks]`

Distinguish between the MLE covariance estimator $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$ and the sample covariance matrix $\mathbf{S}$. State $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}]$ and $\mathbb{E}[\mathbf{S}]$ and explain why the correction factor uses $(n-1)$ instead of $n$.

### ✅ Model Answer

$$\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}})(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})^T \qquad \mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}] = \frac{n-1}{n}\boldsymbol{\Sigma} \quad \text{(biased)}$$

$$\mathbf{S} = \frac{1}{n-1}\sum_{i=1}^n (\mathbf{x}_i - \widehat{\boldsymbol{\mu}})(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})^T \qquad \mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma} \quad \text{(unbiased)}$$

**Why $(n-1)$?** The deviations $(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})$ are measured from the **sample mean** $\widehat{\boldsymbol{\mu}}$, which is itself estimated from the same $n$ observations. This imposes one linear constraint (deviations must sum to zero), consuming **1 degree of freedom**. Only $(n-1)$ deviations are truly free, so dividing by $(n-1)$ corrects the bias. This is **Bessel's correction**.

> **Marking:** 1 mark: both formulae shown. 1 mark: $\mathbb{E}[\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}]$ correct. 1 mark: $\mathbb{E}[\mathbf{S}] = \boldsymbol{\Sigma}$. 1 mark: clear degree-of-freedom explanation.

## Q A5 — Shannon & Differential Entropy `[4 marks]`

**(a)** Write the formula for Shannon entropy $H(X)$ (discrete) and differential entropy $h(X)$ (continuous). **[2]**

**(b)** Explain why $H(X) \geq 0$ always but $h(X)$ can be negative. Give a concrete example where $h(X) < 0$. **[2]**

### ✅ Model Answer

**(a)**
$$H(X) = -\sum_i P(x_i)\ln P(x_i) \qquad \text{range: } [0, \infty)$$
$$h(X) = -\int f(x)\ln f(x)\,dx \qquad \text{range: } (-\infty, \infty)$$

**(b)**
- **Discrete:** Since $0 \leq P(x_i) \leq 1$, we have $\ln P(x_i) \leq 0$, so $-P(x_i)\ln P(x_i) \geq 0$ for each term $\Rightarrow H(X) \geq 0$.
- **Continuous:** The density $f(x)$ can exceed 1 on a narrow interval, making $\ln f(x) > 0$, so the integral can be negative.

**Concrete example:** $X \sim \text{Uniform}[0, a]$ with $a < 1$. Then $f(x) = 1/a > 1$, and:
$$h(X) = \ln(a) < 0 \quad \text{since } a < 1 \Rightarrow \ln a < 0$$

> **Marking:** 1 mark each for correct formulae. 1 mark: clear sign explanation. 1 mark: valid example with $h < 0$.

## Q A6 — Hypothesis Testing `[4 marks]`

Define the **p-value** of a hypothesis test in terms of the test statistic $T$ and the null distribution. State and justify the decision rule for a level $\alpha$ test.

### ✅ Model Answer

For a test where large $T(x)$ gives evidence against $H_0$, and under simple null $H_0: \theta = \theta_0$:

$$\text{p-value:} \quad p(x) = P_{\theta_0}\bigl(T(X) \geq T(x)\bigr)$$

The p-value is the probability, **under $H_0$**, of observing a test statistic at least as extreme as the one actually observed.

**Decision rule:** Reject $H_0$ if and only if $p(x) \leq \alpha$.

**Justification:** $p(x) \leq \alpha \iff T(x) \geq c_\alpha$ (the critical value satisfying $P_{\theta_0}(T \geq c_\alpha) = \alpha$). Rejecting when $p \leq \alpha$ is equivalent to rejecting when $T$ falls in the rejection region $R_\alpha$, which by construction keeps the Type I error $\leq \alpha$.

> **Marking:** 1 mark: correct formula. 1 mark: plain-English meaning. 1 mark: decision rule. 1 mark: valid justification.

## Q B1 — MLE Derivation `[10 marks]`

Starting from the multivariate Gaussian log-likelihood, derive the MLE estimators $\widehat{\boldsymbol{\mu}}_{\text{MLE}}$ and $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}}$.

**(a)** Write the log-likelihood $\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma} \mid \mathbf{X}_{\text{obs}})$ for $n$ iid observations $\mathbf{x}_1, \dots, \mathbf{x}_n \in \mathbb{R}^m$. **[2]**

**(b)** Differentiate with respect to $\boldsymbol{\mu}$ and set to zero to find $\widehat{\boldsymbol{\mu}}_{\text{MLE}}$. **[3]**

**(c)** Differentiate with respect to $\boldsymbol{\Sigma}^{-1}$ using the trace trick and matrix calculus identities, and show $\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n}\sum_i(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})^T$. **[5]**

### ✅ Model Answer

**(a)** The joint log-likelihood (iid assumption → product of densities → sum of log-densities):

$$\ell(\boldsymbol{\mu}, \boldsymbol{\Sigma} \mid \mathbf{X}_{\text{obs}}) = -\frac{nm}{2}\ln(2\pi) - \frac{n}{2}\ln|\boldsymbol{\Sigma}| - \frac{1}{2}\sum_{i=1}^n (\mathbf{x}_i - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1}(\mathbf{x}_i - \boldsymbol{\mu})$$

**(b)** Differentiating the quadratic term with respect to $\boldsymbol{\mu}$:

$$\frac{\partial \ell}{\partial \boldsymbol{\mu}} = \boldsymbol{\Sigma}^{-1}\sum_{i=1}^n(\mathbf{x}_i - \boldsymbol{\mu}) = \mathbf{0}$$

Since $\boldsymbol{\Sigma}^{-1}$ is invertible, this gives $\sum_i(\mathbf{x}_i - \boldsymbol{\mu}) = \mathbf{0}$, hence:

$$\boxed{\widehat{\boldsymbol{\mu}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n \mathbf{x}_i = \bar{\mathbf{x}}}$$

**(c)** Let $\boldsymbol{\Theta} = \boldsymbol{\Sigma}^{-1}$ (precision matrix). Rewrite $\ell$ in terms of $\boldsymbol{\Theta}$:

$$\ell = \text{const} + \frac{n}{2}\ln|\boldsymbol{\Theta}| - \frac{1}{2}\text{Tr}\left[\boldsymbol{\Theta}\sum_i(\mathbf{x}_i-\widehat{\boldsymbol{\mu}})(\mathbf{x}_i-\widehat{\boldsymbol{\mu}})^T\right]$$

Using matrix calculus identities:
- $\frac{\partial}{\partial \boldsymbol{\Theta}}\ln|\boldsymbol{\Theta}| = \boldsymbol{\Theta}^{-1} = \boldsymbol{\Sigma}$
- $\frac{\partial}{\partial \boldsymbol{\Theta}}\text{Tr}[\boldsymbol{\Theta}\mathbf{C}] = \mathbf{C}^T = \mathbf{C}$ (since $\mathbf{C}$ is symmetric)

Setting $\frac{\partial \ell}{\partial \boldsymbol{\Theta}} = \mathbf{0}$:

$$\frac{n}{2}\boldsymbol{\Sigma} - \frac{1}{2}\sum_i(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})^T = \mathbf{0}$$

$$\boxed{\widehat{\boldsymbol{\Sigma}}_{\text{MLE}} = \frac{1}{n}\sum_{i=1}^n(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})(\mathbf{x}_i - \widehat{\boldsymbol{\mu}})^T}$$

## Q B2 — Factor Analysis Derivation `[10 marks]`

In the orthogonal factor model $\mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i$, where $\mathbf{Z}_i \in \mathbb{R}^m$ and $\mathbf{F}_i \in \mathbb{R}^k$:

**(a)** State the three population constraints on $\mathbf{F}_i$ and $\boldsymbol{\epsilon}_i$. **[2]**

**(b)** Derive $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$ by computing $\mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T]$. **[3]**

**(c)** Define communality $h_j^2$ and uniqueness $\varphi_j^2$ and show they partition the unit variance. **[2]**

**(d)** Derive the factor score estimator $\hat{\mathbf{f}}_i = \boldsymbol{\Lambda}^T\mathbf{R}^{-1}\mathbf{z}_i$ using the multivariate Gaussian conditional expectation formula. **[3]**

### ✅ Model Answer

**(a) Three population constraints:**

1. $\mathbb{E}[\mathbf{F}_i] = \mathbf{0}$, $\quad \mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] = \mathbf{I}_k$ (orthonormal common factors)
2. $\mathbb{E}[\boldsymbol{\epsilon}_i] = \mathbf{0}$, $\quad \mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T] = \boldsymbol{\Psi} = \text{diag}(\varphi_1^2, \dots, \varphi_m^2)$ (uncorrelated unique errors)
3. $\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T] = \mathbf{0}_{m \times k}$ (factors orthogonal to errors)

**(b) Deriving the fundamental equation:**

$$\mathbf{R} = \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] = \mathbb{E}[(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)(\boldsymbol{\Lambda}\mathbf{F}_i + \boldsymbol{\epsilon}_i)^T]$$

$$= \boldsymbol{\Lambda}\underbrace{\mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T]}_{=\mathbf{I}}\boldsymbol{\Lambda}^T + \underbrace{\mathbb{E}[\boldsymbol{\epsilon}_i\mathbf{F}_i^T]}_{=\mathbf{0}}\boldsymbol{\Lambda}^T + \boldsymbol{\Lambda}\underbrace{\mathbb{E}[\mathbf{F}_i\boldsymbol{\epsilon}_i^T]}_{=\mathbf{0}} + \underbrace{\mathbb{E}[\boldsymbol{\epsilon}_i\boldsymbol{\epsilon}_i^T]}_{=\boldsymbol{\Psi}}$$

$$\boxed{\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}}$$

**(c) Variance partition:**

The $(j,j)$ diagonal entry of $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$:

$$1 = \underbrace{\sum_{r=1}^k \lambda_{jr}^2}_{h_j^2 \text{ (communality)}} + \underbrace{\varphi_j^2}_{\text{(uniqueness)}}$$

- **Communality** $h_j^2 = \sum_r \lambda_{jr}^2$: variance of feature $j$ explained by the $k$ common factors
- **Uniqueness** $\varphi_j^2 = 1 - h_j^2$: residual/noise variance specific to feature $j$

**(d) Factor score derivation:**

Construct the joint vector $\mathbf{Y}_i = [\mathbf{F}_i^T, \mathbf{Z}_i^T]^T$ with joint covariance:

$$\boldsymbol{\Sigma}_{YY} = \begin{bmatrix} \mathbb{E}[\mathbf{F}_i\mathbf{F}_i^T] & \mathbb{E}[\mathbf{F}_i\mathbf{Z}_i^T] \\ \mathbb{E}[\mathbf{Z}_i\mathbf{F}_i^T] & \mathbb{E}[\mathbf{Z}_i\mathbf{Z}_i^T] \end{bmatrix} = \begin{bmatrix} \mathbf{I} & \boldsymbol{\Lambda}^T \\ \boldsymbol{\Lambda} & \mathbf{R} \end{bmatrix}$$

Apply the Gaussian conditional mean formula $\mathbb{E}[\mathbf{X}_1 | \mathbf{X}_2=\mathbf{x}_2] = \boldsymbol{\mu}_1 + \boldsymbol{\Sigma}_{12}\boldsymbol{\Sigma}_{22}^{-1}(\mathbf{x}_2 - \boldsymbol{\mu}_2)$:

| Term | Role | Value |
|---|---|---|
| $\mathbf{X}_1$ | latent factor | $\mathbf{F}_i$ |
| $\mathbf{X}_2$ | observed | $\mathbf{Z}_i$ |
| $\boldsymbol{\mu}_1, \boldsymbol{\mu}_2$ | means | $\mathbf{0}, \mathbf{0}$ |
| $\boldsymbol{\Sigma}_{12}$ | cross-cov | $\boldsymbol{\Lambda}^T$ |
| $\boldsymbol{\Sigma}_{22}$ | obs cov | $\mathbf{R}$ |

$$\boxed{\hat{\mathbf{f}}_i = \mathbb{E}[\mathbf{F}_i \mid \mathbf{Z}_i = \mathbf{z}_i] = \boldsymbol{\Lambda}^T\mathbf{R}^{-1}\mathbf{z}_i}$$

## Q B3 — MSE Optimality of Conditional Expectation `[10 marks]`

**(a)** Prove that $\mathbb{E}[X \mid Y]$ minimises the mean squared error among all $\sigma(Y)$-measurable estimators $Z$:
$$\mathbb{E}[(X-Z)^2] \geq \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2]$$
**[6]**

**(b)** Using the law of iterated expectations, show that $\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \mathbb{E}[\text{Var}(X \mid Y)]$. **[4]**

### ✅ Model Answer

**(a)** Let $e = X - \mathbb{E}[X \mid Y]$ and $d = \mathbb{E}[X \mid Y] - Z$. Then $X - Z = e + d$, so:

$$\mathbb{E}[(X-Z)^2] = \mathbb{E}[e^2] + 2\mathbb{E}[e \cdot d] + \mathbb{E}[d^2]$$

**Key step — show the cross term is zero:** Since $d = \mathbb{E}[X \mid Y] - Z$ is $\sigma(Y)$-measurable, write $d = g(Y)$. Then:

$$\mathbb{E}[e \cdot d] = \mathbb{E}[(X - \mathbb{E}[X \mid Y]) \cdot g(Y)]$$

$$= \mathbb{E}[X \cdot g(Y)] - \mathbb{E}[\mathbb{E}[X \mid Y] \cdot g(Y)]$$

For the second term, since $g(Y)$ is $\sigma(Y)$-measurable, it can be pulled inside the conditional expectation:

$$\mathbb{E}[\mathbb{E}[X \mid Y] \cdot g(Y)] = \mathbb{E}[\mathbb{E}[X \cdot g(Y) \mid Y]] = \mathbb{E}[X \cdot g(Y)]$$

Therefore $\mathbb{E}[e \cdot d] = 0$, and:

$$\mathbb{E}[(X-Z)^2] = \mathbb{E}[e^2] + \underbrace{\mathbb{E}[d^2]}_{\geq\, 0} \geq \mathbb{E}[e^2] = \mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] \qquad \square$$

Equality holds iff $d = 0$ a.s., i.e., $Z = \mathbb{E}[X \mid Y]$ a.s.

**(b)** Recall $\text{Var}(X \mid Y) = \mathbb{E}[X^2 \mid Y] - (\mathbb{E}[X \mid Y])^2$. Taking expectations:

$$\mathbb{E}[\text{Var}(X \mid Y)] = \mathbb{E}[\mathbb{E}[X^2 \mid Y]] - \mathbb{E}[(\mathbb{E}[X \mid Y])^2] = \mathbb{E}[X^2] - \mathbb{E}[(\mathbb{E}[X \mid Y])^2]$$

Also expand the MSE directly:

$$\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \mathbb{E}[X^2] - 2\mathbb{E}[X \cdot \mathbb{E}[X \mid Y]] + \mathbb{E}[(\mathbb{E}[X \mid Y])^2]$$

The middle term: $\mathbb{E}[X \cdot \mathbb{E}[X \mid Y]] = \mathbb{E}[\mathbb{E}[X \cdot \mathbb{E}[X \mid Y] \mid Y]] = \mathbb{E}[\mathbb{E}[X \mid Y]^2]$

Therefore:

$$\mathbb{E}[(X - \mathbb{E}[X \mid Y])^2] = \mathbb{E}[X^2] - \mathbb{E}[(\mathbb{E}[X \mid Y])^2] = \mathbb{E}[\text{Var}(X \mid Y)] \qquad \square$$

> **Marking:** 6+4. In (a), 3/6 for correct decomposition but incomplete cross-term proof. In (b), 2/4 if total expectation applied but middle term not simplified.

**(a)** Compute the sample mean vector $\widehat{\boldsymbol{\mu}}$. **[2]**

**(b)** Compute the unbiased sample covariance matrix $\mathbf{S}$ using Bessel's correction. Show working for at least two entries. **[4]**

**(c)** Compute the Pearson correlation $r_{12}$ between features 1 and 2. **[2]**

```python
import numpy as np

# Data matrix (rows = observations)
X = np.array([[2,4,1],
              [3,5,2],
              [2,5,1],
              [4,6,3],
              [4,4,2]], dtype=float)

n, m = X.shape

# (a) Sample mean
mu_hat = X.mean(axis=0)
print("(a) Sample mean mu_hat:", mu_hat)

# (b) Sample covariance (ddof=1 applies Bessel's correction)
S = np.cov(X.T, ddof=1)
print("\n(b) Sample covariance matrix S:")
print(np.round(S, 4))

# Manual verification of S_11 and S_12
X_c = X - mu_hat
S11_manual = np.sum(X_c[:,0]**2) / (n-1)
S12_manual = np.sum(X_c[:,0]*X_c[:,1]) / (n-1)
print(f"\nManual check: S_11={S11_manual:.4f}, S_12={S12_manual:.4f}")

# (c) Pearson correlation r_12
r12 = S[0,1] / (np.sqrt(S[0,0]) * np.sqrt(S[1,1]))
print(f"\n(c) Pearson correlation r_12 = {r12:.4f}")
```

## Q C3 — Gaussian Conditioning & Mutual Information `[8 marks]`

A bivariate Gaussian random vector $\mathbf{Z} = [X, Y]^T$ has:
$$\boldsymbol{\mu} = [1.0,\; 2.0]^T, \qquad \boldsymbol{\Sigma} = \begin{bmatrix} 4 & 3 \\ 3 & 9 \end{bmatrix}$$

**(a)** Compute the conditional mean $\mathbb{E}[X \mid Y=5]$ and conditional variance $\text{Var}(X \mid Y=5)$. **[4]**

**(b)** Compute $P(X > 3 \mid Y=5)$. Use $\Phi(0.58) \approx 0.7190$. **[2]**

**(c)** For a bivariate Gaussian with correlation $\rho$, the mutual information is $I(X;Y) = -\frac{1}{2}\ln(1-\rho^2)$. Compute $I(X;Y)$ for the given $\boldsymbol{\Sigma}$. **[2]**

```python
import numpy as np
from scipy import stats

mu_X, mu_Y = 1.0, 2.0
Sigma_XX, Sigma_XY, Sigma_YY = 4.0, 3.0, 9.0

y_obs = 5.0

# (a) Conditional mean and variance
cond_mean = mu_X + Sigma_XY * (1/Sigma_YY) * (y_obs - mu_Y)
cond_var  = Sigma_XX - Sigma_XY * (1/Sigma_YY) * Sigma_XY
print(f"(a) E[X|Y=5]   = {cond_mean:.4f}")
print(f"    Var(X|Y=5) = {cond_var:.4f}")

# (b) P(X > 3 | Y=5)
z_score = (3 - cond_mean) / np.sqrt(cond_var)
prob = 1 - stats.norm.cdf(z_score)
print(f"\n(b) z-score = (3 - {cond_mean}) / sqrt({cond_var}) = {z_score:.4f}")
print(f"    P(X>3|Y=5) = {prob:.4f}")

# (c) Mutual information
rho = Sigma_XY / np.sqrt(Sigma_XX * Sigma_YY)
MI = -0.5 * np.log(1 - rho**2)
print(f"\n(c) rho = {rho:.4f}")
print(f"    I(X;Y) = {MI:.4f} nats")
```

### ✅ Model Answer

**(a)** Using $\mathbb{E}[X \mid Y=y] = \mu_X + \Sigma_{XY}\Sigma_{YY}^{-1}(y - \mu_Y)$:

$$\mathbb{E}[X \mid Y=5] = 1.0 + 3 \cdot \frac{1}{9} \cdot (5-2) = 1.0 + \frac{3}{9}\cdot 3 = 1.0 + 1.0 = \mathbf{2.0}$$

$$\text{Var}(X \mid Y=5) = \Sigma_{XX} - \Sigma_{XY}\Sigma_{YY}^{-1}\Sigma_{YX} = 4 - 3\cdot\frac{1}{9}\cdot 3 = 4 - 1 = \mathbf{3.0}$$

**(b)** $(X \mid Y=5) \sim \mathcal{N}(2.0,\; 3.0)$

$$P(X > 3 \mid Y=5) = P\left(Z > \frac{3-2}{\sqrt{3}}\right) = P(Z > 0.5774) \approx 1 - \Phi(0.58) = 1 - 0.7190 = \mathbf{0.2810}$$

**(c)**
$$\rho = \frac{\Sigma_{XY}}{\sqrt{\Sigma_{XX}\Sigma_{YY}}} = \frac{3}{\sqrt{4 \times 9}} = \frac{3}{6} = 0.5$$

$$I(X;Y) = -\frac{1}{2}\ln(1 - 0.25) = -\frac{1}{2}\ln(0.75) = -\frac{1}{2}(-0.2877) = \mathbf{0.1438 \text{ nats}}$$

> **Marking:** 4+2+2. In (a), 2 marks for mean, 2 for variance. In (b), accept 0.279–0.283 range. In (c), must compute $\rho$ explicitly.

## Q C4 — Factor Analysis: Communality & Factor Scores `[8 marks]`

A factor model has $m=3$ variables, $k=1$ factor:
$$\widehat{\boldsymbol{\lambda}} = [0.9,\; 0.7,\; 0.5]^T, \qquad \widehat{\boldsymbol{\Psi}} = \text{diag}(0.19,\; 0.51,\; 0.75)$$

**(a)** Verify the fundamental equation $\mathbf{R} = \boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}$ for entries $(1,1)$ and $(1,2)$. **[3]**

**(b)** Compute communality $h_j^2$ and uniqueness $\varphi_j^2$ for all three variables. Which is best explained by the common factor? **[2]**

**(c)** For $\mathbf{z}_i = [1.2,\; 0.8,\; 0.6]^T$, compute the factor score $\hat{f}_i = \boldsymbol{\lambda}^T\mathbf{R}^{-1}\mathbf{z}_i$.
Use: $\mathbf{R}^{-1} = \begin{bmatrix}2.48 & -1.52 & -0.40 \\ -1.52 & 2.31 & 0.26 \\ -0.40 & 0.26 & 1.42\end{bmatrix}$ (given). **[3]**

```python
import numpy as np

lam = np.array([0.9, 0.7, 0.5])  # k=1, so Lambda is a vector
Psi = np.diag([0.19, 0.51, 0.75])

# (a) Fundamental equation
LLT = np.outer(lam, lam)  # Lambda Lambda^T for k=1
R = LLT + Psi
print("(a) Lambda Lambda^T:")
print(np.round(LLT, 4))
print("\nR = Lambda Lambda^T + Psi:")
print(np.round(R, 4))
print(f"\nVerify R[1,1] = {R[0,0]:.4f}  (should be 1.0)")
print(f"Verify R[1,2] = {R[0,1]:.4f}  (= lambda_1 * lambda_2 = {lam[0]*lam[1]:.4f})")

# (b) Communality and uniqueness
h2 = lam**2  # since k=1
phi2 = 1 - h2
print("\n(b) Communalities and uniquenesses:")
for j in range(3):
    print(f"  Variable {j+1}: h²={h2[j]:.4f}, phi²={phi2[j]:.4f}")
print(f"  Best explained: Variable {np.argmax(h2)+1} (h²={h2.max():.4f})")

# (c) Factor score
R_inv = np.array([[2.48, -1.52, -0.40],
                  [-1.52, 2.31, 0.26],
                  [-0.40, 0.26, 1.42]])
z_i = np.array([1.2, 0.8, 0.6])

Rinv_z = R_inv @ z_i
f_hat = lam @ Rinv_z
print(f"\n(c) R_inv @ z_i = {np.round(Rinv_z, 4)}")
print(f"    Factor score f_hat = lambda^T (R_inv z_i) = {f_hat:.4f}")
```

### ✅ Model Answer

**(a)** For $k=1$: $(\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T)_{ij} = \lambda_i \lambda_j$

- Entry $(1,1)$: $[\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}]_{11} = 0.9^2 + 0.19 = 0.81 + 0.19 = \mathbf{1.00}$ ✓ (diagonal of $\mathbf{R}$ must equal 1)
- Entry $(1,2)$: $[\boldsymbol{\Lambda}\boldsymbol{\Lambda}^T + \boldsymbol{\Psi}]_{12} = 0.9 \times 0.7 + 0 = \mathbf{0.63}$ ✓

**(b)**

| Variable | $h_j^2 = \lambda_j^2$ | $\varphi_j^2 = 1 - h_j^2$ |
|---|---|---|
| 1 | $0.81$ | $0.19$ |
| 2 | $0.49$ | $0.51$ |
| 3 | $0.25$ | $0.75$ |

**Variable 1** is best explained ($h_1^2 = 0.81$).

**(c)**
$$\mathbf{R}^{-1}\mathbf{z}_i = \begin{bmatrix}2.48(1.2)+(-1.52)(0.8)+(-0.40)(0.6)\\(-1.52)(1.2)+2.31(0.8)+0.26(0.6)\\(-0.40)(1.2)+0.26(0.8)+1.42(0.6)\end{bmatrix} = \begin{bmatrix}1.520\\0.180\\0.580\end{bmatrix}$$

$$\hat{f}_i = \boldsymbol{\lambda}^T(\mathbf{R}^{-1}\mathbf{z}_i) = 0.9(1.520)+0.7(0.180)+0.5(0.580) = 1.368+0.126+0.290 = \mathbf{1.7840}$$

> **Marking:** 3+2+3. In (c), must show $\mathbf{R}^{-1}\mathbf{z}_i$ step explicitly before the inner product.

## Q D1 — PCA Control Limits in SHM `[7 marks]`

A SHM system monitors a bridge using $m=10$ sensors. PCA is trained on $n=500$ healthy baseline observations. The 99% control limits are: $T^2_{\text{lim}} = 14.2$ and $Q_{\text{lim}} = 0.05$.

**(a)** A new observation gives $T^2 = 18.7$ and $Q = 0.003$. Classify the structural state and explain. **[3]**

**(b)** The next observation gives $T^2 = 8.1$ and $Q = 0.21$. Interpret. What physical phenomenon causes large $Q$ but normal $T^2$? **[2]**

**(c)** Why is Slutsky's theorem needed to operationalise the CLT for constructing these control limits? **[2]**

### ✅ Model Answer

**(a)** $T^2 = 18.7 > 14.2$ (exceeds limit) but $Q = 0.003 < 0.05$ (within limit).

The observation has an unusual position **within** the modelled PC subspace, but its residual energy is normal. The observation lies inside the trained correlation structure (no new mode) but at an extreme point. This could indicate **heavy operational loading** (e.g., unusual traffic density) rather than structural damage. **Recommended action:** flag for investigation, not immediate alarm.

**(b)** $T^2 = 8.1 < 14.2$ (normal) but $Q = 0.21 > 0.05$ (exceeds limit).

The observation projects normally within the $k$-PC subspace but has large residual energy outside it. **Physical cause:** a new damage mechanism (e.g., a crack altering the inter-sensor correlation structure) that is **not captured by the baseline loading matrix** $\widehat{\mathbf{P}}_k$. **(c)** The CLT guarantees $\widehat{\boldsymbol{\mu}}_n \sim \mathcal{N}(\boldsymbol{\mu}, \frac{1}{n}\boldsymbol{\Sigma})$ asymptotically, but $\boldsymbol{\Sigma}$ is **unknown** in practice. Slutsky's theorem states that if $\mathbf{S} \xrightarrow{p} \boldsymbol{\Sigma}$, then replacing $\boldsymbol{\Sigma}$ with $\mathbf{S}$ does not change the limiting distribution. This justifies using $\mathbf{S}$ to build control limits without knowing the true population covariance.

## Q D2 — FA Loading Matrix Interpretation `[7 marks]`

FA with $k=2$ factors on $m=6$ SHM features gives the loading matrix:

| Feature | $F_1$ | $F_2$ |
|---|---|---|
| Strain-L1 | 0.85 | 0.10 |
| Accel-L1 | 0.80 | 0.15 |
| Strain-L2 | 0.05 | 0.88 |
| Accel-L2 | 0.10 | 0.82 |
| Temp-L1 | 0.42 | 0.35 |
| Temp-L2 | 0.38 | 0.40 |

**(a)** Interpret the physical meaning of $F_1$ and $F_2$. **[2]**

**(b)** Compute $h_j^2$ and $\varphi_j^2$ for Strain-L1 and Temp-L1. Which carries more unique variance and why? **[3]**

**(c)** If $\boldsymbol{\Psi} \to \mathbf{0}$ (uniquenesses vanish), what happens to $I(\mathbf{Z}; \mathbf{F})$ and what does this mean physically? **[2]**

```python
import numpy as np

# Loading matrix (rows: features, cols: factors)
Lambda = np.array([
    [0.85, 0.10],  # Strain-L1
    [0.80, 0.15],  # Accel-L1
    [0.05, 0.88],  # Strain-L2
    [0.10, 0.82],  # Accel-L2
    [0.42, 0.35],  # Temp-L1
    [0.38, 0.40],  # Temp-L2
])

features = ['Strain-L1','Accel-L1','Strain-L2','Accel-L2','Temp-L1','Temp-L2']

# (b) Communalities and uniquenesses
h2 = np.sum(Lambda**2, axis=1)
phi2 = 1 - h2

print("Communalities and Uniquenesses:")
for i, f in enumerate(features):
    print(f"  {f:12s}: h²={h2[i]:.4f}, phi²={phi2[i]:.4f}")
```

### ✅ Model Answer

**(a)**
- **$F_1$** loads heavily on Strain-L1 (0.85) and Accel-L1 (0.80) but negligibly on Location-2 features. $F_1$ represents the **structural dynamic response at Location 1** (local vibration mode or load path).
- **$F_2$** loads heavily on Strain-L2 (0.88) and Accel-L2 (0.82), representing the **structural response at Location 2**.
- Temperature features load moderately on both factors — reflecting **global thermal effects** shared across both locations.

**(b)**
$$\text{Strain-L1:} \quad h^2 = 0.85^2 + 0.10^2 = 0.7225 + 0.0100 = 0.7325, \quad \varphi^2 = 0.2675$$
$$\text{Temp-L1:} \quad h^2 = 0.42^2 + 0.35^2 = 0.1764 + 0.1225 = 0.2989, \quad \varphi^2 = 0.7011$$

**Temp-L1 carries far more unique variance** ($\varphi^2 = 0.7011$). Physical reason: temperature is driven by ambient conditions (solar radiation, wind, seasonal cycles) that are **not related to structural vibration modes** captured by $F_1$ and $F_2$.

**(c)** As $\boldsymbol{\Psi} \to \mathbf{0}$, unique errors vanish and $\mathbf{Z}_i = \boldsymbol{\Lambda}\mathbf{F}_i$ exactly (deterministic mapping). Knowing $\mathbf{Z}_i$ gives **perfect knowledge** of $\mathbf{F}_i$, so $I(\mathbf{Z}; \mathbf{F}) \to \infty$. Physically: the features provide infinite information about the latent structural state — an idealised scenario. Real sensors always have non-zero noise ($\boldsymbol{\Psi} \neq \mathbf{0}$).

> **Marking:** 2+3+2. In (a), location-specific physical interpretation required. In (b), 1 mark per communality, 1 mark identifying Temp-L1 with physical reason.
